```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('./data/data_officesupplies.csv')
df.shape
```

```python
df.head()
```

```python
# 세련되지 않은 방법
df_west = df[df['Region']=='West'].Units.reset_index(drop=True)
df_central = df[df['Region']=='Central'].Units.reset_index(drop=True)
df_east = df[df['Region']=='East'].Units.reset_index(drop=True)
```

```python
plt.boxplot([df_west, df_central, df_east], labels=['West','Central','East'])
plt.show()
```

```python
df2 = pd.DataFrame(data={'West':list(df_west)}) 
# west east central 각각 길이가 달라서 dataframe으로는 
# boxplot을 하나 밖에 못그린다
df2.plot.box()
plt.show()
```

```python
my_group = df.groupby('Region')
my_list = []
my_names = []
```

```python
for name, dfb in my_group :
    # groupby를 하면 튜플로 되고
    # for 문으로 돌릴때 첫번째 원소는 이름들이고
    # 두번째 원소는 이름으로 묶여진 dataFrame이다.
    my_names.append(name)
    my_list.append(dfb.Units)
```

```python
plt.boxplot(my_list, labels=my_names)
plt.show()
```
