```python
plt.hist(np.log(sr+1), bins=50, color='green', alpha=0.4, density=True)
# np.log 전처리를 해주면서 정규분포의 그래프로 만들어 줄 수 있다.
# 그리고 이 그래프에서 매장의 개수가 0인것은 인위적으로 결측치를 0으로 만들었다고 볼 수 있으므로
# 이런것은 직접 보면서 알아내 처리를 해줄 필요가 있다.
plt.show()
```
